```python
#Mission 08
score = int(input("Enter your score: "))
def grade(score):
    if score >= 80:
        print("4")
    elif score >= 70 and score < 80:
        print("3")
    elif score >= 60 and score < 70:
        print("2")
    elif score >= 50 and score < 60:
        print("1")
    elif score < 50:
        print("0")
def vldtscore(score):
    if score < 0 or score > 100:
        print("Error")
    else:
        grade(score)
vldtscore(score)
```

```python
#Mission 10
def triangle(base, height):
    area = (base * height) / 2
    print("The area of the triangle is: ", area)

def square(side):
    area = (side * side)
    print("The area of the square is: ", area)

def rectangle(length, width):
    area = (length * width)
    print("The area of the rectangle is: ", area)

selmenu = input("Choose menu with A (Triangle), B (Square) or C (Rectangle): ")
if selmenu == "A" or selmenu == "a":
    print("You selected A (Triangle)")
    height = int((input("Enter the height of the triangle: ")))
    base = int((input("Enter the base of the triangle: ")))
    triangle(base, height)
elif selmenu == "B" or selmenu == "b":
    print("You selected B (Square)")
    side = int((input("Enter the side of the square: ")))
    square(side)
elif selmenu == "C" or selmenu == "c":
    print("You selected C (Rectangle)")
    length = int((input("Enter the length of the rectangle: ")))
    width = int((input("Enter the width of the rectangle: ")))
    rectangle(length, width)
else :
    print("Error")
```
